## Theory

### 1. Univariate Data
- **Definition**: Univariate data involves **only one variable**. Analysis focuses on describing its distribution, central tendency, and spread.  
- **Key Points**:
  - Single variable focus → No relationships studied.  
  - Used for descriptive analysis → Mean, median, variance, etc.  
  - Visualization → Histograms, bar plots, boxplots.  

**Examples**  
- Heights of students in a class.  
- Salaries of workers in a company.  
- Number of defects in manufactured products.  

**Techniques**

| Technique  | Python Function/Library | Purpose |
|------------|--------------------------|---------|
| Histogram  | `sns.histplot()`, `plt.hist()` | Distribution of a numeric variable |
| Box Plot   | `sns.boxplot()` | Spread, median, outliers |
| Violin Plot| `sns.violinplot()` | Density + distribution |
| KDE Plot   | `sns.kdeplot()` | Smooth distribution curve |
| Bar Plot   | `sns.barplot()`, `plt.bar()` | Categorical frequency |
| Pie Chart  | `plt.pie()` | Proportions (less recommended) |

---

### 2. Bivariate Data
- **Definition**: Bivariate data contains **two variables** studied together. Focus is on **relationships, correlations, and dependencies**.  
- **Key Points**:
  - Independent (X) and Dependent (Y) variables may exist.  
  - Correlation ranges from -1 to 1 (negative, positive, or no relation).  
  - Regression is often used for prediction.  

**Examples**  
- Ice cream sales vs. temperature.  
- Height vs. weight of individuals.  
- Income vs. education level.

|------------|--------------------------|---------|

##  **Types of Analysis We Perform**

### ➤ **Univariate Analysis** – Understanding One Feature
In univariate analysis, we study one variable at a time to describe its distribution, central tendency, and spread.

**Example Questions:**
- How is the distribution of the total bill amount?
- Are there outliers or extreme values in the data?

**Techniques Used:**
- **Histogram**: A bar chart showing how frequently different ranges of values occur.
- **Boxplot**: A visual summary of data distribution including median, quartiles, and outliers.

---

### ➤ **Bivariate Analysis** – Exploring Relationships Between Two Features
This type of analysis helps us see how two variables interact.

**Example Questions:**
- Does the size of the bill influence the tip amount?
- How does the bill amount vary across different days?

**Techniques Used:**
- **Scatter Plot**: Shows the relationship between two continuous variables.
- **Boxplot by Category**: Compares one variable across different groups (e.g., total bill by day).

---

### ➤ **Multivariate Analysis** – Analyzing More Than Two Features
Multivariate analysis provides deeper insights by considering three or more variables together. This helps in understanding complex relationships and dependencies.

**Example Questions:**
- How do bill amounts and tips differ between men and women during lunch and dinner?
- What are the correlations between numeric features like bill amount, tip, and party size?

**Techniques Used:**
- **Pair Plot**: Displays scatter plots between every pair of numeric variables and helps to find patterns.
- **Correlation Heatmap**: Shows how strongly variables are related to each other using correlation coefficients.
- **Facet Grid**: Splits the data by categories and creates multiple subplots for comparison.

---

## **Why Visualizations Are Important**

Visualizations help by:
- Making data easier to understand at a glance.
- Showing patterns that are not obvious from raw numbers.
- Highlighting areas that need further investigation.
- Communicating findings to others effectively.

For instance, a **scatter plot** between `total_bill` and `tip` can reveal whether larger bills tend to get larger tips, which can be useful for restaurant management.

###  **Correlation**
Correlation is a measure that tells us how strongly two numeric variables are related. Values range from:
- `1` → Perfect positive correlation.
- `0` → No correlation.
- `-1` → Perfect negative correlation.

For example, if `total_bill` and `tip` have a correlation close to `1`, it means as the bill increases, the tip amount also increases.

```python
# Importing necessary libraries
import pandas as pd  # pandas is used for data manipulation and analysis, especially with tables (DataFrames).
import seaborn as sns  # seaborn is used for creating attractive and informative statistical graphics easily.
import matplotlib.pyplot as plt  # matplotlib is a plotting library used to display and customize graphs.

```

```python
# -------------------------------
# 1. DATA COLLECTION
# -------------------------------

# Load the dataset from the CSV file located at "C:\MITADT\ISDL\Assignment5\tips.csv"
# Using 'r' before the path tells Python to treat backslashes as literal characters.
tips = pd.read_csv(r"C:\MITADT\ISDL\Assignment5\tips.csv")

# 'tips' is now a DataFrame, which is like a table where you can access, filter, and manipulate data easily.
# For example, tips.head() will show the first 5 rows of the dataset.

```

```python
# -------------------------------
# UNIVARIATE ANALYSIS (Analyzing one variable at a time)
# -------------------------------

# HISTOGRAM
# This plot shows how often each range of 'total_bill' values appears in the data.
# The 'kde=True' argument adds a smooth curve to represent the distribution.
sns.histplot(tips['total_bill'], kde=True)
plt.title('Histogram of Total Bill')  # Adds a title to the graph.
plt.show()  # Displays the plot.

# BOXPLOT
# This plot helps visualize the distribution of 'total_bill', showing the median, quartiles, and outliers.
sns.boxplot(x=tips['total_bill'])
plt.title('Boxplot of Total Bill')  # Adds a title to the graph.
plt.show()  # Displays the plot.

```

```python
# -------------------------------
# BIVARIATE ANALYSIS (Analyzing the relationship between two variables)
# -------------------------------

# SCATTER PLOT
# This plot shows the relationship between 'total_bill' and 'tip'.
# Each point represents one observation in the dataset.
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')  # Adds a title to the graph.
plt.show()  # Displays the plot.

# BOXPLOT BY CATEGORY
# This boxplot compares the distribution of 'total_bill' across different days of the week.
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot of Total Bill by Day')  # Adds a title to the graph.
plt.show()  # Displays the plot.

```

```python
# -------------------------------
# MULTIVARIATE ANALYSIS (Analyzing three or more variables together)
# -------------------------------

# PAIR PLOT WITH HUE
# This creates scatterplots for every pair of numeric columns in the dataset.
# 'hue' colors the data points by 'sex' to visually distinguish groups.
sns.pairplot(tips, hue='sex')
plt.show()  # Displays the plot.

# CORRELATION HEATMAP
# 'corr()' calculates correlation coefficients between numeric variables (values between -1 and 1).
# 'heatmap' shows these correlations with colors, where annot=True displays the numbers.
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')  # 'coolwarm' makes high and low values easily distinguishable.
plt.title('Correlation Heatmap')  # Adds a title to the graph.
plt.show()  # Displays the plot.

# FACET GRID
# This divides the dataset by 'time' and 'sex' and creates separate scatter plots for each subgroup.
# 'map' applies the scatterplot function to each group.
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")  # Plots 'total_bill' vs 'tip' in each subplot.
plt.show()  # Displays all the subplots.
```
