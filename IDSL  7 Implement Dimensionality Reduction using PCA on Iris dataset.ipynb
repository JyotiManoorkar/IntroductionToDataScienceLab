## Methods of Dimensionality Reduction

---

### A. Linear Methods

1. **PCA (Principal Component Analysis)**  
   - Projects data onto directions of **maximum variance**.

2. **LDA (Linear Discriminant Analysis)**  
   - Maximizes **separation between classes**.  
   - Supervised method (uses class labels).

3. **Factor Analysis**  
   - Finds **latent variables** explaining correlations between features.

---

### B. Non-linear / Manifold Learning Methods

1. **t-SNE (t-distributed Stochastic Neighbor Embedding)**  
   - Preserves **local structure** for visualization in 2D/3D.

2. **Isomap**  
   - Preserves **global geometric structure** of the data.

3. **UMAP (Uniform Manifold Approximation and Projection)**  
   - Faster alternative to t-SNE, **scalable** for larger datasets.

4. **Kernel PCA**  
   - Captures **non-linear structures** using kernels.

---

### C. Feature Selection Methods

- Instead of transforming features, these methods **select the most important ones**:  
  - **Variance Thresholding**  
  - **Recursive Feature Elimination (RFE)**  
  - **SelectKBest / Mutual Information**


# Understanding PCA (Principal Component Analysis)

---

## 1. What is a Dimension?
- Each feature in a dataset = one **dimension**.  
- Example (Iris dataset has 4 features):  
  - Sepal length  
  - Sepal width  
  - Petal length  
  - Petal width  

üëâ Each flower is like a **point in 4D space** (since it has 4 numbers).  

**Analogy:**  
- Moving on a line ‚Üí **1D**  
- Moving on a plane ‚Üí **2D**  
- Moving in a room ‚Üí **3D**  
- Extra features add extra directions (**4D is hard to imagine**).

---

## 2. Why Reduce Dimensions?
- Humans can only visualize up to **3D**.  
- Too many features can be **confusing** and may **repeat information**.  
- Reducing dimensions helps us:  
  - Visualize data in **2D or 3D**  
  - **Remove redundancy** (e.g., petal length & width are correlated)  
  - Detect **patterns and clusters**  
  - Make **machine learning faster and simpler**  

**Analogy:** A messy desk becomes easier to understand if you **stack and organize papers**. PCA does the same with data.

---

## 3. What is PCA?
- **PCA = Principal Component Analysis**  
- A method to **reduce dimensions** while keeping the most **important information**.  
- It creates **new axes (directions)** called **Principal Components (PCs)**.


Concept:

    PCA finds new axes (called principal components) that:

    Capture maximum variance in the data.

    Are orthogonal (uncorrelated).

    The first principal component captures the most variance, the second captures the next most, and so on.

Simple analogy:

    ‚ÄúIf you have a cloud of points in 3D, PCA finds the direction where the cloud is stretched the most ‚Äî that‚Äôs your first principal component.‚Äù
    

Mathematical intuition :

    Center the data.

    Compute the covariance matrix.

    Compute its eigenvectors and eigenvalues.

    Sort eigenvectors by eigenvalues ‚Äî they define new axes.

    Project the data onto top k components.






---

## 4. Principal Components
- **PC1 (First Principal Component):** The direction with the **most variation**.  
- **PC2 (Second Principal Component):** The direction with the **second-most variation**, always **perpendicular to PC1**.  
- **PC3, PC4, ‚Ä¶:** Capture smaller variations, each orthogonal to the previous PCs.  

‚ÄúPCA stands for Principal Component Analysis. It‚Äôs a statistical method that transforms correlated features into a set of uncorrelated components. Each principal component captures as much variance as possible.
Think of it as rotating your data to find the directions where it‚Äôs most spread out ‚Äî those are the axes that matter most. We can drop the rest with minimal information loss.‚Äù

**Analogy:** PC1 is like the **longest shadow of a group of points**, PC2 is the **next longest shadow at a right angle**.

---

## 5. Why PCA is Useful
- Makes data **simpler** (fewer dimensions).  
- Removes **noise or redundant information**.  
- Lets us **see patterns clearly** in 2D/3D plots.  
- Helps create **new features** from existing ones.  


### What PCA Does:

Identifies new axes (principal components) along directions of maximum variance.

These new axes are uncorrelated (orthogonal).

It‚Äôs an unsupervised technique (doesn‚Äôt use labels).

‚ÄúPCA rotates the coordinate system to align with the directions that best represent your data.
Think of it as finding the most ‚Äòinteresting‚Äô directions in your dataset.‚Äù

 ### Step-by-Step Explanation:

Standardize the data (mean = 0, variance = 1).

Compute covariance matrix ‚Üí measures relationships between features.

Find eigenvectors and eigenvalues:

Eigenvectors ‚Üí directions (principal components).

Eigenvalues ‚Üí magnitude (variance captured).

Sort eigenvectors by descending eigenvalues.

Project data onto top k components.

‚ÄúWe can think of PCA as finding the best line, plane, or hyperplane that preserves the structure of our data with minimal loss.‚Äù


---

## 6. Why Use PCA on the Iris Dataset

### 6.1 The Problem
- Iris dataset has **4 features per flower**:  
  - Sepal length, Sepal width, Petal length, Petal width  
- Each flower = point in **4D space**, impossible to visualize.  
- Some features are **similar or correlated**, e.g., petal length & width.  
  - Longer petals are usually wider ‚Üí **repeated information**.  

---

### 6.2 How PCA Helps
- PCA finds **best new directions** (principal components) to view the data.  
- These directions capture **most of the variation** while ignoring redundancy.  
- Think of it like **tilting and looking at a messy garden from the best angle**:  
  - **PC1:** Direction where flowers are most spread out  
  - **PC2:** Next best direction at a right angle to PC1  

- Plotting flowers along PC1 and PC2:  
  - **Setosa** ‚Üí separate cluster  
  - **Versicolor & Virginica** ‚Üí partially separated  
- Patterns hidden in 4D now become **visible in 2D**.

---

### 6.3 Simple Analogy
- A **pile of toys on the floor**:  
  - Look from the **best angle** to see groups clearly ‚Üí PCA does the same for data.  
- Two rulers measuring the same thing ‚Üí redundant  
  - PCA combines these into **one principal component**, removing repeated info.

---

### ‚úÖ In Short
- PCA reduces **4 features ‚Üí 2 or 3 principal components**.  
- Helps to:  
  - **Visualize clusters** of flowers  
  - **Remove redundant information**  
  - **Reveal hidden patterns** clearly  
  - **Simplify machine learning tasks**


```python





# Import necessary libraries
import matplotlib.pyplot as plt                    # For plotting 2D and 3D graphs
from sklearn.datasets import load_iris            # To load the Iris dataset
from sklearn.preprocessing import StandardScaler  # For standardizing features
from sklearn.decomposition import PCA             # For Principal Component Analysis (dimensionality reduction)

# Step 1: Load the Iris dataset
iris = load_iris()            # Load the built-in Iris dataset
X = iris.data                 # Feature matrix (150 samples x 4 features: sepal/petal length & width)
y = iris.target               # Labels array (0, 1, 2) representing three species
target_names = iris.target_names  # Names of species ("setosa", "versicolor", "virginica")

print("Original Data Shape:", X.shape)  # Check shape of dataset (150, 4)

# Step 2: Standardize the features
# PCA works best when features have similar scales
scaler = StandardScaler()        # Create scaler object
X_scaled = scaler.fit_transform(X)  # Standardize data: (value - mean)/std deviation

# Step 3: Apply PCA to reduce 4D data to 2D
pca = PCA(n_components=2)         # Keep top 2 principal components

# ‚ÄúNow, we initialize PCA with n_components=2, meaning we‚Äôll reduce our 4-dimensional data into 2 dimensions.
#The fit_transform() method fits the model and applies the dimensionality reduction in one step.
# Behind the scenes, this uses the covariance matrix and eigen decomposition.‚Äù





X_pca = pca.fit_transform(X_scaled)  # Fit PCA and transform the data

# Step 4: Plot the 2D PCA results
plt.figure(figsize=(8,6))         # Set figure size
colors = ['r', 'g', 'b']          # Red, Green, Blue for three species
# Loop through each species and plot its points
for target, color, label in zip([0, 1, 2], colors, target_names):
    plt.scatter(X_pca[y == target, 0],   # PC1 values for the species
                X_pca[y == target, 1],   # PC2 values for the species
                color=color,             # Color for the species
                label=label,             # Label for legend
                alpha=0.7,               # Transparency
                edgecolors='k')          # Black edges for better visibility

plt.xlabel("Principal Component 1")      # Label X-axis
plt.ylabel("Principal Component 2")      # Label Y-axis
plt.title("PCA on Iris Dataset (2D)")    # Plot title
plt.legend()                              # Show legend
plt.grid(True)                             # Show grid lines
plt.show()                                 # Display the plot

# Step 5: Print explained variance
# Shows the amount of information (variance) captured by each principal component
print("Explained variance ratio:", pca.explained_variance_ratio_)  # Fraction of variance per PC
print("Total variance retained:", sum(pca.explained_variance_ratio_))  # Total variance retained (~96%)
```
