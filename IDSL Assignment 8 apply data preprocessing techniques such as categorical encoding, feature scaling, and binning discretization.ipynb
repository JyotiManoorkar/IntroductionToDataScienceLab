# üìò Data Preprocessing Techniques ‚Äì Detailed Notes

Data preprocessing is the first and most crucial step in the data analysis or machine learning workflow.  
It involves cleaning, transforming, and organizing raw data into a usable format for modeling.  

This note covers **three major preprocessing techniques**:
1. Categorical Encoding  
2. Feature Scaling  
3. Binning / Discretization  

---

## üß© 1Ô∏è‚É£ Categorical Encoding

### üîπ Definition
Categorical data are variables that contain label values rather than numeric values.  
Machine learning models cannot process text directly, so **categorical encoding** converts these categories into numeric representations.

---

### üîπ Why Encoding Is Needed
- Most ML algorithms (like regression, SVM, neural networks) require numeric input.  
- Encoding converts **nominal** or **ordinal** text data into numbers.  
- Prevents the model from misinterpreting text as string data.

---

### üîπ Types of Categorical Encoding

#### **A. Label Encoding**
- Converts each unique category into a unique integer.
- Formula:  
  Category ‚Üí Integer Mapping  
  Example:  
  | Fuel Type | Encoded Value |  
  |------------|---------------|  
  | Petrol     | 2 |  
  | Diesel     | 0 |  
  | Electric   | 1 |

**Advantages:**
- Simple and memory efficient.  
- Suitable for ordinal data (where order matters).

**Disadvantages:**
- Introduces false ordering for nominal data (e.g., Red=1, Blue=2 doesn‚Äôt mean Red < Blue).

---

#### **B. One-Hot Encoding**
- Creates new binary columns (0 or 1) for each category.
- Example:  
  | Fuel_Type_Petrol | Fuel_Type_Diesel | Fuel_Type_Electric |  
  |------------------|------------------|--------------------|  
  | 1 | 0 | 0 |  
  | 0 | 1 | 0 |  
  | 0 | 0 | 1 |

**Advantages:**
- Removes false ordinal relationships.  
- Widely used for nominal (unordered) data.

**Disadvantages:**
- Increases the number of columns (dimensionality).

---

#### **C. Target / Frequency Encoding (Advanced)**
- Replaces each category with the mean of the target variable or frequency count.
- Useful for large categorical variables in ML.

---

### ‚úÖ **When to Use**
- Use **Label Encoding** for **ordered categories** (e.g., Low, Medium, High).  
- Use **One-Hot Encoding** for **unordered categories** (e.g., Brand, Fuel Type).

---

## ‚öñÔ∏è 2Ô∏è‚É£ Feature Scaling

### üîπ Definition
Feature scaling is the process of bringing all features to the same scale or range.  
It ensures that no single feature dominates others due to its magnitude.

---

### üîπ Why Scaling Is Needed
- Many algorithms (like Gradient Descent, KNN, SVM, PCA) are **distance-based**.  
- If one feature has larger numeric values, it will dominate distance calculations.  
- Scaling improves **model convergence** and **accuracy**.

---

### üîπ Common Methods of Scaling

#### **A. Min-Max Scaling (Normalization)**
- Rescales the feature to a fixed range [0, 1].
- Formula:  
  \[
  X' = \frac{X - X_{min}}{X_{max} - X_{min}}
  \]

**Example:**
If mileage ranges from 10 to 30 km/l,  
and a car has 20 km/l,  
\[
X' = \frac{20 - 10}{30 - 10} = 0.5
\]

**Use Case:** Neural Networks, KNN, Clustering.

---

#### **B. Standardization (Z-score Scaling)**
- Centers data around mean 0 and standard deviation 1.
- Formula:  
  \[
  X' = \frac{X - \mu}{\sigma}
  \]

**Example:**  
If mean price = 25, std = 10, and car price = 35,  
\[
X' = \frac{35 - 25}{10} = 1.0
\]

**Use Case:** Regression, SVM, PCA, algorithms assuming normal distribution.

---

#### **C. Robust Scaling**
- Uses median and IQR (Interquartile Range), less sensitive to outliers.
- Formula:  
  \[
  X' = \frac{X - median(X)}{IQR(X)}
  \]

**Use Case:** Datasets with outliers.

---

### ‚úÖ **When to Use**
- Use **Min-Max** when data has a defined boundary.  
- Use **Standardization** when data follows normal distribution.  
- Use **Robust Scaling** when data has outliers.

## üìä 3Ô∏è‚É£ Binning / Discretization

### üîπ Definition
Binning (also called discretization) is the process of converting **continuous features** into **categorical bins** or intervals.

---

### üîπ Why Binning Is Needed
- Reduces noise or small observation errors.  
- Converts continuous features into discrete categories for easier interpretation.  
- Helps capture non-linear relationships.  
- Makes models like Decision Trees easier to understand.

---

### üîπ Types of Binning

#### **A. Equal-Width Binning**
- Divides the entire range of data into equal intervals.  
- Each bin has the same width but may contain different numbers of observations.

**Example:**

- Low (‚Çπ0‚Äì‚Çπ17L)  
- Medium (‚Çπ17‚Äì‚Çπ34L)  
- High (‚Çπ34‚Äì‚Çπ50L)

# demonstrates data preprocessing techniques ‚Äî encoding categorical variables, scaling numerical data, and binning continuous values ‚Äî using a small car dataset.

## üßæ Summary of the Script

### **Overall Goal**
This script demonstrates key **data preprocessing techniques** ‚Äî encoding categorical variables, scaling numerical data, and binning continuous values ‚Äî using a small car dataset.

---

### **Step 1: Load / Create Dataset**
A sample dataset is created with car details:
- **Categorical columns:** `Brand`, `Fuel_Type`, `Transmission`
- **Numerical columns:** `Mileage(km/l)`, `Price(Lakhs)`

This serves as the base `DataFrame` for all transformations.

---

### **Step 2: Categorical Encoding**
Transforms text-based categories into numeric values:
- **Label Encoding (`Fuel_Type`)** ‚Üí assigns numeric codes to fuel types.
- **One-Hot Encoding (`Transmission`)** ‚Üí creates binary columns such as `Trans_Manual` and `Trans_Automatic`.

**Purpose:** Makes categorical data suitable for machine learning algorithms.

---

### **Step 3: Feature Scaling**
Normalizes numeric data for consistent scale:
- **Min-Max Scaling (`Mileage`)** ‚Üí rescales values between 0 and 1.
- **Standardization (`Price`)** ‚Üí centers data with mean = 0 and standard deviation = 1.

**Purpose:** Prevents features with larger numeric ranges from dominating model training.

---

### **Step 4: Binning / Discretization**
Converts continuous values into categories:
- **Equal-width binning (`Price`)** ‚Üí splits prices into *Low*, *Medium*, and *High* ranges.
- **Equal-frequency binning (`Mileage`)** ‚Üí divides mileage into bins with equal sample counts.

**Purpose:** Simplifies continuous data and highlights distribution patterns.

---

### **Step 5: Visualization**
Displays a **bar chart** of price categories using equal-width binning.

**Purpose:** Helps visualize how many cars fall into each price range.

---



````python
# üîπ Step 1: Load / Create Dataset
# We are manually creating a dataset of cars to demonstrate preprocessing steps.
# This dataset includes both:
#   1. Categorical features: text-based data (e.g., 'Brand', 'Fuel_Type', 'Transmission')
#   2. Numerical features: numbers (e.g., 'Mileage(km/l)', 'Price(Lakhs)')
# Why: Having a mix allows us to show how to handle both types, as real-world data usually contains both.
# Creating a small dataset manually also makes it easier to understand and verify each preprocessing step.

import pandas as pd   # For creating and manipulating tabular data (DataFrames)
import numpy as np    # For numerical operations, arrays, and mathematical calculations
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt   # For creating visualizations to understand data

# Create dataset using a dictionary: keys are column names, values are lists of column data
data = {
    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Honda', 'Mahindra', 'Toyota', 'Ford'],  # Car brands
    'Fuel_Type': ['Petrol', 'Diesel', 'Electric', 'Petrol', 'Diesel', 'Petrol', 'Diesel'],  # Type of fuel used
    'Transmission': ['Manual', 'Automatic', 'Manual', 'Automatic', 'Manual', 'Automatic', 'Manual'],  # Gear type
    'Mileage(km/l)': [18, 20, 0, 22, 15, 19, 17],  # How far the car goes per liter of fuel
    'Price(Lakhs)': [5.5, 8.2, 12, 7.5, 6.8, 9.5, 8.0]  # Car price in lakhs
}

# Convert dictionary into a pandas DataFrame
# Why: DataFrames are like spreadsheets in Python; they make data manipulation easy.
df = pd.DataFrame(data)
print("Original Dataset:\n", df)


# üîπ Step 2: Categorical Encoding
# Why: Machine learning models cannot work directly with text. Algorithms expect numerical input.
# Example: Computers cannot compare "Petrol" and "Diesel" directly; we need to convert these categories into numbers.

# 1. Label Encoding
# Assigns a unique integer to each category in a column.
# Fuel_Type mapping here: Diesel -> 0, Electric -> 1, Petrol -> 2
# Why: Simple and efficient for columns with a small number of categories.
# Caution: Label encoding can introduce a sense of order. Only use when categories are ordinal or model handles non-ordinal numeric categories well.
le = LabelEncoder()
df['Fuel_Type_Encoded'] = le.fit_transform(df['Fuel_Type'])

# 2. One-Hot Encoding
# Creates separate binary columns for each category.
# Transmission column will become:
#   Trans_Manual: 1 if car is manual, 0 otherwise
#   Trans_Automatic: 1 if car is automatic, 0 otherwise
# Why: Prevents models from assuming any numeric order between categories.
df = pd.get_dummies(df, columns=['Transmission'], prefix='Trans')

print("\nAfter Encoding:\n", df)


# üîπ Step 3: Feature Scaling
# Why: Features like 'Mileage' and 'Price' are on different scales. Scaling ensures that:
# - No feature dominates due to its magnitude
# - Models like KNN, SVM, gradient descent-based models converge faster
# Two common scaling methods are used here:
# 1. Min-Max Scaling: Scales values to a 0-1 range
# 2. Standardization: Centers values around mean 0 with standard deviation 1

# Min-Max Scaling for Mileage
# Converts mileage into 0-1 range based on min and max mileage
# Why: Preserves the shape of the distribution while bringing all values into the same range.
scaler_mileage = MinMaxScaler()
df['Mileage_Scaled'] = scaler_mileage.fit_transform(df[['Mileage(km/l)']])

# Standardization for Price
# Adjusts price so that it has mean 0 and standard deviation 1
# Why: Useful when feature distribution varies greatly; helps algorithms like linear regression and gradient descent converge better.
scaler_price = StandardScaler()
df['Price_Standardized'] = scaler_price.fit_transform(df[['Price(Lakhs)']])

print("\nAfter Scaling:\n", df)


# üîπ Step 4: Binning / Discretization
# Binning groups continuous numerical values into discrete categories.
# Why: Makes analysis easier, reduces noise, and helps algorithms handle non-linear relationships.

# Equal-width binning for Price
# Splits the price range into 3 equal intervals and labels them Low, Medium, High
# Why: Easy to interpret; helps to quickly categorize cars by price.
df['Price_Bin'] = pd.cut(df['Price(Lakhs)'], bins=3, labels=['Low', 'Medium', 'High'])

# Equal-frequency binning for Mileage
# Splits the mileage data so each bin contains approximately the same number of cars
# Why: Useful when data is skewed; ensures each bin has sufficient samples for analysis.
df['Mileage_Bin'] = pd.qcut(df['Mileage(km/l)'], q=3, labels=['Low', 'Medium', 'High'])

print("\nAfter Binning:\n", df)


# üîπ Step 5: Visualization
# Visualizing data helps to understand distributions, patterns, and trends.
# Here we plot how many cars fall into each price category.
# Why: Visual inspection can often reveal insights that raw numbers cannot.

# Count number of cars in each price bin
price_bin_counts = df['Price_Bin'].value_counts().sort_index()

# Create a bar chart
# kind='bar': vertical bars
# color='skyblue', edgecolor='black': aesthetic choices for readability
plt.figure(figsize=(6,4))
price_bin_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Cars in Each Price Bin')  # Chart title
plt.xlabel('Price Category')  # X-axis label
plt.ylabel('Number of Cars')  # Y-axis label
plt.show()
````
